### Reading an image using Open CV

```python
from matplotlib import pyplot as plt
import cv2

# Read the image
image = cv2.imread('C:/Users/asus/Desktop/Tuwaiq Academy/Dr. Afshan/horse.jpg')
if image is None:
    print("Error: Could not load image.")
else:
    # Convert BGR to RGB for proper display in Matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.axis('off')  # Turn off axes
    plt.show()
```

### Converting the image to gray scale 

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('C:/Users/asus/Desktop/Tuwaiq Academy/Dr. Afshan/horse.jpg')  # Replace 'image.jpg' with the path to your image

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image using Matplotlib
plt.figure(figsize=(6, 6))
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Hide axes for better visualization
plt.show()
```

#### Edge Detection using canny algorithm 

```python
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('C:/Users/asus/Desktop/Tuwaiq Academy/Dr. Afshan/horse.jpg', cv2.IMREAD_GRAYSCALE)

# Apply the Canny edge detection algorithm
edges = cv2.Canny(image, 100, 200)

# Display the edges using matplotlib
plt.figure(figsize=(8, 8))
plt.imshow(edges, cmap='gray')
plt.title('Edges Detected')
plt.axis('off')  # Hide the axes
plt.show()
```
